```python
import numpy as np
import matplotlib.pyplot as plt

def load_data(n_samples=100):
    X = np.linspace(-1.0, 2.0, n_samples).reshape(-1, 1)
    noise = np.random.randn(n_samples).reshape(-1, 1) * 0.15
    y = X**2 -X + 0.25 + noise
    return X, y

def plot_dataset(X, y):
    plt.scatter(X, y, s=15, c='k')
    plt.xlabel(r'$x$', fontsize=12)
    plt.ylabel(r'$y$', rotation=True, fontsize=12)

def plot_prediction(regressor, axes, n_samples=100):
    x = np.linspace(axes[0], axes[1], n_samples).reshape(-1, 1)
    y_pred = regressor.predict(x)
    plt.plot(x, y_pred, color='y', label=r'$\^y$', linewidth=2.5)
    plt.xlim(axes[0], axes[1])
    plt.legend()
```

```python
from sklearn.tree import DecisionTreeRegressor

X, y = load_data()

tree_reg = DecisionTreeRegressor()
tree_reg.fit(X, y)

tree_reg_leaf_10 = DecisionTreeRegressor(min_samples_leaf=10)
tree_reg_leaf_10.fit(X, y)

axes = [-1.5, 2.5]
plt.figure('Tree regression', figsize=(14,4))

plt.subplot(121)
plt.title('No restriction')
plot_prediction(tree_reg, axes)
plot_dataset(X, y)

plt.subplot(122)
plt.title('min_samples_leaf={}'.format(10))
plot_prediction(tree_reg_leaf_10, axes)
plot_dataset(X ,y)

plt.show()
```
