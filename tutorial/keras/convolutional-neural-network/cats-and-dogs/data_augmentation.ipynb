```python
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os

dataset_dir = '/home/guhwanbae/workspace/machine-learning/resource/cats-and-dogs-small/'
train_cats_dir = os.path.join(dataset_dir, 'train/cats')

file_names = sorted([os.path.join(train_cats_dir, file_name)
    for file_name in os.listdir(train_cats_dir)])

idx = np.random.randint(len(file_names))
sample_path = file_names[idx]
img = image.load_img(sample_path, target_size=(150, 150))
print('Random index =', idx)

plt.figure('Original image')
plt.title('Original image')
plt.imshow(img)

img = np.asarray(img)
img = img.reshape((1,) + img.shape)
```

```python
datagen = image.ImageDataGenerator(rotation_range=20,
                                   width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   shear_range=0.1,
                                   zoom_range=0.1,
                                   horizontal_flip=True,
                                   fill_mode='nearest',
                                   rescale=1./255)

for idx, batch in enumerate(datagen.flow(img, batch_size=1)):
    title = 'iteration =%d' % idx
    plt.figure(title)
    plt.title(title)
    plt.imshow(batch[0])
    if idx > 3:
        break
```
