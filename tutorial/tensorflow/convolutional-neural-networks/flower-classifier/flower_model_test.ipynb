```python
import flower_model

flower_model = flower_model.FlowerModel()
flower_model.fetch()
labels = flower_model.get_class_labels()
print('Flower labels: {}'.format(labels))
```

```python
import matplotlib.pyplot as plt

for flower_class in labels:
    n_samples_per_class = 3
    samples = flower_model.get_random_samples(flower_class, n_samples_per_class)
    plt.figure(figsize=(7,4))
    for idx, sample_image in enumerate(samples):
        fig_idx = 100 + 10*n_samples_per_class + (idx+1)
        plt.subplot(fig_idx)
        plt.title('class: {}'.format(flower_class))
        plt.imshow(sample_image)
        plt.axis('off')
    plt.show()
```
