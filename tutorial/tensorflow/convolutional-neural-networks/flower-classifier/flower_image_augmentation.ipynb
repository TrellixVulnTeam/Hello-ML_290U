```python
import numpy as np
import matplotlib.pyplot as plt
from flower_model import FlowerModel
import image_transform

flower_model = FlowerModel()
flower_model.fetch()
labels = flower_model.get_class_labels()

print('class: {}'.format(labels))
```

```python
def augmentation_samples(images, n_repeat):
    aug_samples = np.repeat(images, n_repeat, axis=0)
    return np.asarray([image_transform.random_transform(sample) for sample in aug_samples])
```

```python
n_samples = 10
n_repeat = 5

samples = flower_model.get_random_samples('daisy', n_samples, resize_shape=(299,299))
outputs = augmentation_samples(samples, n_repeat)

outputs_by_sample = outputs.reshape(n_samples, n_repeat, 299, 299, 3)

for original_sample, aug_samples in zip(samples, outputs_by_sample):
    plt.figure(figsize=(15,5))
    plt.subplot(100 + 10*(n_repeat+1) + 1)
    plt.title('Original')
    plt.imshow(original_sample)
    for i, aug_sample in enumerate(aug_samples):
        subplot_idx = 100 + 10*(n_repeat+1) + (2+i)
        plt.subplot(subplot_idx)
        plt.title('Augmentation: {}'.format(i+1))
        plt.imshow(aug_sample)
    plt.show()
```
